```python
# imports
from conllu import parse
```

```python
# read the CoNLL-U file
with open("fi_tdt-ud-train.conllu", "r", encoding="utf-8") as f:
    data = f.read()

# parse the CoNLL-U data
sentences = parse(data)

# initialize dictionary to store different forms of verbs
verb_form_dict = {}

# counting the amounts of different forms of verbs
for sentence in sentences: 
    for token in sentence:
        if token["upos"] == "VERB":
            if token["lemma"] not in verb_form_dict:
                verb_form_dict[token["lemma"]] = [token["form"]]
            elif token["form"] not in verb_form_dict[token["lemma"]]:
                verb_form_dict[token["lemma"]].append(token["form"])
print(verb_form_dict)
```
